```python
def normalize_price(dataframe,population):
    specific_population = population

    filtered_ds = dataframe[dataframe['level5'] == specific_population]
    min_price = filtered_ds['price'].min()
    max_price = filtered_ds['price'].max()
    return (filtered_ds['price'] - min_price) / (max_price - min_price)


fig,ax = plt.subplots() 
ax.hist([normalize_price(sub_ds,"Fuenlabrada")], label="Fuenlabrada", histtype="step", density=True)
ax.hist([normalize_price(sub_ds,"Leganés")],label="Leganés", histtype="step", density=True)
ax.hist([normalize_price(sub_ds,"Getafe")],label="Getafe", histtype="step", density=True)
ax.hist([normalize_price(sub_ds,"Alcorcón")],label="Alcorcón", histtype="step", density=True)
#, normalize_price(sub_ds,"Leganés"), normalize_price(sub_ds,"Getafe"), normalize_price(sub_ds,"Alcorcón")],  label=['Fuenlabrada', 'Leganés', 'Getafe','Alcorcón'])
plt.legend(loc='upper right')
plt.show()
```

#### Exercise 21. Make a plot of the coordinates (latitude and longitude columns) of the south belt of Madrid by color of each population (you must use the subset obtained in the question 14) (★★★★)

Execute the following cell and then start coding in the next one. You must implement a simple code that transform the coordinates columns in a Python dictionary (add more information if needed) and then add it to the map

```python
coordinates_dict = {}

for index, row in sub_ds.iterrows():
    latitude = float(row['latitude'].replace(',', '.'))
    longitude = float(row['longitude'].replace(',', '.'))
    coordinates_dict[index] = (latitude, longitude)


from ipyleaflet import Map, Marker


map_center = [40.28, -3.7]  
mymap = Map(center=map_center, zoom=10)


for index, coordinates in coordinates_dict.items():
    latitude, longitude = coordinates
    marker = Marker(location=(latitude, longitude), draggable=False, title=str(index))  
    mymap.add_layer(marker)


display(mymap)
```
