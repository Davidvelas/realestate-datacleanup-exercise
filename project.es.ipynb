```python
import pandas as pd

import matplotlib.pyplot as plt
def mean_price_population(dataframe,population):
    
    
    specific_population = population

    filtered_ds = dataframe[dataframe['level5'] == specific_population]

    mean_price = filtered_ds['price'].mean()
    
    return mean_price


labels = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]
values = [mean_price_population(sub_ds,"Fuenlabrada"), mean_price_population(sub_ds,"Leganés"), mean_price_population(sub_ds,"Getafe"), mean_price_population(sub_ds,"Alcorcón")]

plt.figure(figsize = (10, 5))

plt.bar(labels, values)

plt.title("Mean prices")
plt.show()
```

```python
coordinates_dict = {}

for index, row in sub_ds.iterrows():
    latitude = float(row['latitude'].replace(',', '.'))
    longitude = float(row['longitude'].replace(',', '.'))
    coordinates_dict[index] = (latitude, longitude)

print(coordinates_dict)
from ipyleaflet import Map, Marker

# Create a map centered at a specific location
map_center = [40.28, -3.7]  # Set your desired center coordinates
mymap = Map(center=map_center, zoom=10)

# Iterate over the coordinates dictionary and add markers to the map
for index, coordinates in coordinates_dict.items():
    latitude, longitude = coordinates
    marker = Marker(location=(latitude, longitude), draggable=False, title=str(index))  # Convert index to string
    mymap.add_layer(marker)

# Display the map
display(mymap)
```
